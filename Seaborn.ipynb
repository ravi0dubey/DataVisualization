```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tps = sns.load_dataset('tips')
tps
```

```python
#Plotting graph using relplot

sns.relplot(x='tip' ,y='total_bill', data =tps)
```

```python
#Plotting graph using relplot, showing data based on sex

sns.relplot(x='tip' ,y='total_bill', data =tps, hue = 'sex')
```

```python
#Plotting graph using relplot, showing data based on day of the week

sns.relplot(x='tip' ,y='total_bill', data =tps, hue = 'day')
```

```python
# Plotting graph using relplot, showing data based on day of the week and 
# seggregating based on type of people(Smoker and non smoker)

sns.relplot(x='tip' ,y='total_bill', data =tps, hue = 'day', style = 'smoker')
```

```python
# Plotting graph using relplot, showing data based on day of the week and 
# seggregating based on type of people(Smoker and non smoker)
# data plot size is based on the sex

sns.relplot(x='tip' ,y='total_bill', data =tps, hue = 'day', style = 'smoker', size = 'sex')
```

```python
# Plotting Line graph using relplot, showing data based on day of the week and 


sns.relplot(x='tip' ,y='total_bill', data =tps, hue = 'day', kind = 'line')
```

```python
# creating  line plot for entire dataset


df1 = pd.DataFrame(np.random.randint(4,100,(100,5)) , columns = ['col1','col2', 'col3', 'col4', 'col5'])
df1

sns.relplot(data= df1, kind = 'line')
```

```python
# creating a categorical plot based on categorical data, day of the week and numerical data = 'total bill'

sns.catplot(x= 'day', y = 'total_bill', data = tps)
```

```python
# creating a categorical plot based on categorical data, smoker and numerical data = 'tip'

sns.catplot(x= 'smoker', y = 'tip', data = tps)
```

```python
# creating a box plot based on categorical data, day of the week and numerical data = 'tip'

sns.catplot( data = tps , x= 'day', y = 'tip', kind = 'box', hue = 'sex')
```

```python
# creating  line plot using col1 and col2


df1 = pd.DataFrame(np.random.randint(4,100,(100,5)) , columns = ['col1','col2', 'col3', 'col4', 'col5'])
df1

sns.relplot(x= 'col1', y = 'col2', data= df1, kind = 'line')
```

```python
# creating  line plot using col1 and col2


df1 = pd.DataFrame(np.random.randint(4,100,(100,5)) , columns = ['col1','col2', 'col3', 'col4', 'col5'])
df1

sns.relplot(x= 'col1', y = 'col2', data= df1, kind = 'line', sort = True)
```
